```python
# My First ML Project!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("ðŸš€ My ML environment is working!")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")

# Let's create some fun data
np.random.seed(42)  # So we get the same results each time

# Simulate test scores vs hours studied
hours_studied = np.random.randint(1, 11, 50)  # 1-10 hours
test_scores = hours_studied * 8 + np.random.normal(0, 10, 50)  # Some noise
test_scores = np.clip(test_scores, 0, 100)  # Keep scores between 0-100

# Create a DataFrame (like Excel but in Python)
df = pd.DataFrame({
    'hours_studied': hours_studied,
    'test_score': test_scores
})

print("\nFirst 10 rows of our data:")
print(df.head(10))

# Make a cool visualization
plt.figure(figsize=(10, 6))
plt.scatter(df['hours_studied'], df['test_score'], alpha=0.7, s=60)
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Test Scores vs Hours Studied')
plt.grid(True, alpha=0.3)
plt.show()

# Basic statistics
print(f"\nAverage test score: {df['test_score'].mean():.1f}")
print(f"Students who studied 8+ hours averaged: {df[df['hours_studied'] >= 8]['test_score'].mean():.1f}")
print(f"Students who studied <4 hours averaged: {df[df['hours_studied'] < 4]['test_score'].mean():.1f}")
```
